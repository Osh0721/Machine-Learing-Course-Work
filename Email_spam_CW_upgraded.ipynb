```python
upper_limit_total = new_data["capital_run_length_total"].mean() + 3*new_data["capital_run_length_total"].std()
lower_limit_total = new_data["capital_run_length_total"].mean() - 3*new_data["capital_run_length_total"].std()

upper_limit_longest = new_data["capital_run_length_longest"].mean() + 3*new_data["capital_run_length_longest"].std()
lower_limit_longest = new_data["capital_run_length_longest"].mean() - 3*new_data["capital_run_length_longest"].std()

print("Upper limit_total",upper_limit_total)
print("lower limit total",lower_limit_total)

print("Upper limit  longest",upper_limit_longest)
print("lower limit longest ",lower_limit_longest)
```

```python
outliers = (new_data["capital_run_length_total"] > upper_limit_total) | (new_data["capital_run_length_total"] < lower_limit_total) | (new_data["capital_run_length_longest"] > upper_limit_longest) | (new_data["capital_run_length_longest"] < lower_limit_longest)
```

```python
new_data_no_outlier = new_data.loc[~outliers]
```

```python
X = new_data_no_outlier.iloc[:, :57].values
Y = new_data_no_outlier.iloc[:, 57].values
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.20, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
#KNN without pca
from sklearn.neighbors import KNeighborsClassifier

classifier = KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2)
classifier.fit(X_train,Y_train)

Y_pred = classifier.predict(X_test)

from sklearn.metrics import confusion_matrix,accuracy_score

print(accuracy_score(Y_test,Y_pred))

# Compute confusion matrix for KNN
cm_knn = confusion_matrix(Y_test, Y_pred)
# Plot confusion matrix heatmap for KNN
plt.figure()
sns.heatmap(cm_knn, annot=True, fmt='g', cmap='Blues', square=True)
plt.xlabel('Predicted label')
plt.ylabel('True label')
plt.title('Confusion Matrix for KNN')
plt.show()
from sklearn.metrics import classification_report

print(classification_report(Y_test, Y_pred))
```

```python
#Decision Tree without PCA
from sklearn.tree import  DecisionTreeClassifier
classifer=DecisionTreeClassifier(criterion='entropy', random_state = 0)
classifer.fit(X_train,Y_train)

Y_pred=classifer.predict(X_test)
from sklearn.metrics import confusion_matrix,accuracy_score

print(accuracy_score(Y_test,Y_pred))

# Compute confusion matrix for Decision Tree
cm_dt = confusion_matrix(Y_test, Y_pred)

# Plot confusion matrix heatmap for Decision Tree
plt.figure()
sns.heatmap(cm_dt, annot=True, fmt='g', cmap='Blues', square=True)
plt.xlabel('Predicted label')
plt.ylabel('True label')
plt.title('Confusion Matrix for Decision Tree')
plt.show()

from sklearn.metrics import classification_report

print(classification_report(Y_test, Y_pred))
```
